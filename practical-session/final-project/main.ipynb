# Time Series Visualization: Number of Songs Released Per Month

This section visualizes the number of songs released per month as a time series using the 'created_at' column.


```python
# Time Series Plot: Number of Songs Released Per Month using 'created_at'
if 'created_at' in df.columns:
    df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
    df = df.dropna(subset=['created_at'])
    df['release_month'] = df['created_at'].dt.to_period('M').astype(str)
    songs_per_month = df['release_month'].value_counts().sort_index()
    plt.figure(figsize=(16, 6))
    sns.lineplot(x=songs_per_month.index, y=songs_per_month.values, marker='o')
    plt.title('Number of Songs Released Per Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Songs')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No 'created_at' column found. Please check your dataframe columns and update the code if needed.")
```
